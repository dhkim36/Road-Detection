```python
import os
import cv2
import numpy as np

from tqdm.notebook import tqdm
from matplotlib import pyplot as plt
```

```python
mask_dir = '/home/dhkim/work/RoadDetector/albu-solution/data/Contec/train/Masks_512_road'
```

```python
mask_list = os.listdir(mask_dir)

for mask_file in tqdm(mask_list[:]):
    mask_path = os.path.join(mask_dir, mask_file)
    
    ori_img = cv2.imread(mask_path)
    gray_img = cv2.cvtColor(ori_img, cv2.COLOR_RGB2GRAY)
    c_gray_img = np.where(gray_img == 179, 255, 0)
    
    cv2.imwrite(mask_path, c_gray_img)
```
